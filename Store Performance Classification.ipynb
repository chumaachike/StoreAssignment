```python
def detect_outliers(data):
    outliers = []
    threshold = 3
    mean = np.mean(data)
    std =  np.std(data)
    for i in data:
        z_score = (i-mean)/std
        if np.abs(z_score) >threshold:
            outliers.append(i)
    return outliers
```

```python
def calculate_metrics(confusion_matrix):
    """
    Calculates accuracy, precision, recall, F1 score, and false positive rate from a given confusion matrix.
    
    Args:
    confusion_matrix (numpy array): A 2x2 numpy array representing the confusion matrix, where the rows correspond to
                                     the actual class labels and the columns correspond to the predicted class labels.
    
    Returns:
    A dictionary containing the following metrics:
        - accuracy
        - precision
        - recall
        - f1_score
        - false_positive_rate
    """
    
    tp = confusion_matrix[0, 0]
    fp = confusion_matrix[0, 1]
    fn = confusion_matrix[1, 0]
    tn = confusion_matrix[1, 1]
    
    # Calculate accuracy
    accuracy = (tp + tn) / (tp + fp + fn + tn)
    
    # Calculate precision
    precision = tp / (tp + fp)
    
    # Calculate recall
    recall = tp / (tp + fn)
    
    # Calculate F1 score
    f1_score = 2 * precision * recall / (precision + recall)
    
    # Calculate false positive rate
    false_positive_rate = fp / (fp + tn)
    
    metrics = {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1_score,
        'false_positive_rate': false_positive_rate
    }
    
    return metrics
```

### Data Cleaning 

```python
#get a copy of the dataset which would be cleaneed 
train_cleaned = train.copy()
# Select only rows where the country is UK
train_cleaned = train_cleaned[(train_cleaned.Country =="UK")]
# Remove outliers in staff
train_cleaned = train_cleaned[(train_cleaned['Staff'] < 299) & (train_cleaned['Staff'] > 0)]
train_cleaned = train_cleaned[(train_cleaned['30 min population'] < 1929088)]
train_cleaned = train_cleaned[(train_cleaned['20 min population'] < 1723394)]
train_cleaned = train_cleaned[(train_cleaned['10 min population'] < 1397709)]
```

```python
train_cleaned = train_cleaned.drop(['Town', 'Store ID', 'Manager name'], axis=1)
```

From the description of the dataset we can see that there are four unique values for the car park variable out of a posible two. The next step is to plot a bar chart and see what other posiblities are there.


```python
train_cleaned["Car park"].value_counts().plot(kind="bar")
plt.show()
```

From the histogram we can see that there are some variables that where wrongly encoded as Y instead of Yes and N instead of No. We need to fix that by  converting all the Y to Yes and N to No

```python
# we can use a dictionary to set up a mapping that would replace the values Y and N with Yes and No respectively
train_cleaned['Car park'] = train_cleaned['Car park'].replace({'Y': 'Yes', 'N':'No'})
#confirm that it worked
train_cleaned['Car park'].value_counts().plot(kind='bar')
```

```python
train_cleaned['Performance'] = train_cleaned['Performance'].replace({'Good': 1, 'Bad':0})
```

```python
# to check which of variables contain unique values all through so as to exclude them and avoid zero variance predictors we use the unique function in Numpy
for i in range(train_cleaned.shape[1]):
    length_of_unique_data = len(np.unique(train_cleaned.iloc[:,i].values))
    unique_percentage = (length_of_unique_data/train_cleaned.shape[0]) * 100
    print (f"the uniqueness of data in index {i} is {unique_percentage}")
```

```python
train_cleaned = train_cleaned.drop('Country', axis=1)
```

```python
from sklearn.preprocessing import OneHotEncoder
# one hot encoding
from sklearn.compose import make_column_transformer
transformer = make_column_transformer(
    (OneHotEncoder(handle_unknown='ignore'), ['Car park', 'Location']),
    remainder='passthrough')
```

```python
train_cleaned_enc = transformer.fit_transform(train_cleaned)
train_cleaned = pd.DataFrame(train_cleaned_enc, columns=transformer.get_feature_names_out())
```

```python
train_cleaned = train_cleaned.reset_index(drop=True)
```

```python
train_cleaned = train_cleaned.drop(['onehotencoder__Car park_No', 'onehotencoder__Location_Retail Park'], axis=1)
```

```python
train_cleaned.columns
```

```python
selected_features = ['onehotencoder__Car park_Yes', 'onehotencoder__Location_High Street',
       'onehotencoder__Location_Shopping Centre', 'remainder__Staff',
       'remainder__Floor Space', 'remainder__Window',
       'remainder__Demographic score', 'remainder__40min population',
       'remainder__30 min population', 'remainder__20 min population',
       'remainder__10 min population', 'remainder__Store age',
       'remainder__Clearance space', 'remainder__Competition number',
       'remainder__Competition score', 'remainder__Performance']
```

```python
train_cleaned = train_cleaned[selected_features]
```

```python
targetName = "remainder__Performance"
train_cleaned.columns.get_loc(targetName)
featureNames=train_cleaned.columns[np.where(train_cleaned.columns != targetName)]
X_train = train_cleaned[featureNames]
y_train = train_cleaned[targetName]
train_cleaned['remainder__Performance'].value_counts().plot(kind='bar')
```

```python
from sklearn.preprocessing import MinMaxScaler
mm_scaler = MinMaxScaler()
X_train = pd.DataFrame(mm_scaler.fit_transform(X_train), columns=X_train.columns)
```

```python
test['Performance'] = test['Performance'].replace({'Good': 1, 'Bad':0})
test['Car park'] = test['Car park'].replace({'Y': 'Yes', 'N':'No'})
test = test[(test['Staff'] > 0)]
test = test.reset_index(drop=True)
test_enc = transformer.transform(test)
test = pd.DataFrame(test_enc, columns=transformer.get_feature_names_out())
X_test = test[featureNames]
y_test = test[targetName]
X_test = pd.DataFrame(mm_scaler.transform(X_test), columns=X_test.columns)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import log_loss
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
param_grid = {'C': np.arange(0,10,0.2),
              'penalty': ['l1', 'l2'],
              'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag'],
              'max_iter': [100, 500, 1000],
             }
```

```python
log_reg = LogisticRegression(random_state=31415)
grid_search = GridSearchCV(log_reg, param_grid, cv=5, scoring='precision')
```

```python
grid_search.fit(X_train, y_train)
```

```python
print('Best hyperparameters: ', grid_search.best_params_)
print('Best cross-entropy loss: {:.2f}'.format(grid_search.best_score_))
```

```python
log_reg = LogisticRegression(solver='newton-cg', C=0.8, penalty='l2', max_iter=100, random_state=31415).fit(X_train, y_train)
```

```python
y_pred_log_reg = log_reg.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred_log_reg)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```

```python
metrics = calculate_metrics(cm)
print(metrics)
```
